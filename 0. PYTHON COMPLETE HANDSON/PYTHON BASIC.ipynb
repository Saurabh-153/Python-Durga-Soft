## Fundamental Data Types vs Immutability

**Note:**

- All fundamental data types in Python are immutable, which means that once an object is created, it cannot be modified. If we attempt to modify the object, a new object will be created with the changes applied. This behavior is known as immutability.

- In Python, when a new object is required, the Python Virtual Machine (PVM) does not immediately create a new object. Instead, it first checks if an object with the required content is already available. If an object with the same content exists, it will be reused. Only if such an object is not available will a new object be created. This approach improves memory utilization and performance.

- However, the issue with this approach arises when multiple references point to the same object. If one reference is allowed to change the content of the existing object, it would impact the remaining references. To address this, the concept of immutability is introduced. Once an object is created, its content cannot be changed. If we attempt to modify the object, a new object will be created with the desired changes.

The immutability of fundamental data types in Python helps ensure data consistency and prevents unintended modifications in multi-reference scenarios.


````python
a = 10
b = 10
a is b              
print(id(a))               
print(id(b))   
````

**Explanation:**

In the code snippet provided:

- `a = 10` assigns the value 10 to the variable `a`.
- `b = 10` assigns the value 10 to the variable `b`.
- `a is b` checks if `a` and `b` refer to the same object. In this case, it returns `True`, indicating that `a` and `b` are referencing the same object in memory.
- `id(a)` returns the unique identifier of the object that `a` is referencing, which is `1572353952`.
- `id(b)` returns the unique identifier of the object that `b` is referencing, which is also `1572353952`.

This behavior occurs because small integers (typically between -5 and 256) are cached by Python. When two variables are assigned the same small integer value, they will point to the same object in memory, resulting in the same unique identifier.

Please note that this behavior may vary for larger integers or other types of objects, where Python may create separate objects for each assignment even if their values are the same.

## List in Python

A list is a data structure that can hold a collection of objects. The order of the objects in a list is preserved, and duplicates are allowed. Lists are dynamic, meaning that they can grow or shrink as needed.

**Characteristics of Lists**

* **Insertion order is preserved.** The order in which the objects are added to the list is the order in which they will be retrieved.
* **Duplicate objects are allowed.** You can add the same object to a list multiple times.
* **Heterogeneous objects are allowed.** A list can contain objects of different types.
* **Lists are dynamic.** You can add or remove objects from a list at any time.
* **Elements are placed within square brackets and with comma separator.** For example, a list of numbers could be `[10, 20, 30]`.
* **We can differentiate duplicate elements by using index.** The index of an object in a list is its position in the list, starting from 0. For example, the index of the number 20 in the list `[10, 20, 30]` is 1.
* **We can preserve insertion order by using index.** The index is also used to preserve the insertion order of the objects in a list. For example, if you add the number 40 to the list `[10, 20, 30]` at the index 2, the list will become `[10, 20, 40, 30]`.
* **Python supports both positive and negative indexes.** Positive indexes start from 0, and negative indexes start from -1. For example, the negative index of the number 20 in the list `[10, 20, 30]` is -1.
* **List objects are mutable.** This means that we can change the content of a list after it has been created. For example, we can add, remove, or replace objects in a list.

### Aliasing and Cloning of List Objects

The process of giving another reference variable to the existing list is called aliasing.

````python
x=[10,20,30,40]
y=x 
print(id(x)) 
print(id(y))
````

````python
x = [10,20,30,40]
y = x
y[1] = 777
print(x)
````

The problem in this approach is by using one reference variable if we are changing content, then those changes will be reflected to the other reference variable <br>

To overcome this problem we should go for cloning. The process of creating exactly duplicate independent object is called cloning.<br>

We can implement cloning by using slice operator or by using copy() function.

````python
# By using Slice Operator
x = [10,20,30,40]
y = x[:]
y[1] = 777
print(x)
print(y)
````

````python
# By using copy() Function
x = [10,20,30,40]
y = x.copy()
y[1] = 777
print(x)
print(y)
````

### Mathematical Operations on Sets

The following mathematical operations can be performed on sets:

* **Union:** Returns a new set that contains all the elements of the two sets.
    * `x.union(y)` or `x | y`
* **Intersection:** Returns a new set that contains only the elements that are common to the two sets.
    * `x.intersection(y)` or `x & y`
* **Difference:** Returns a new set that contains all the elements that are in the first set but not in the second set.
    * `x.difference(y)` or `x - y`
* **Symmetric difference:** Returns a new set that contains all the elements that are in either of the two sets but not in both sets.
    * `x.symmetric_difference(y)` or `x ^ y`

Below are some examples of how to use these operations:

````python
#python
x = {10, 20, 30, 40}
y = {30, 40, 50, 60}

print(x.union(y))  # {10, 20, 30, 40, 50, 60}
print(x | y)  # {10, 20, 30, 40, 50, 60}

print(x.intersection(y))  # {40, 30}
print(x & y)  # {40, 30}

print(x.difference(y))  # {10, 20}
print(x - y)  # {10, 20}

print(y.difference(x))  # {50, 60}

print(x.symmetric_difference(y))  # {10, 50, 20, 60}
print(x ^ y)  # {10, 50, 20, 60}
````

## Dictionary In Python

* **Advantages of using dictionaries:**
    * Duplicate keys are not allowed but values can be duplicated.
    * Heterogeneous objects are allowed for both key and values.
    * Insertion order is not preserved.
    * Dictionaries are mutable.
    * Dictionaries are dynamic.
    * Indexing and slicing concepts are not applicable.

    * ipaddress: domain name

```python
dictionary = {

### How to Delete Elements from a Dictionary in Python

There are three ways to delete elements from a dictionary in Python:

### Important Functions of Dictionary in Python

Dictionaries are a data structure that can be used to store key-value pairs. Keys must be unique, but values can be duplicated.

Here are some of the important functions of dictionary in Python:

* **`dict()`:** This function is used to create a dictionary.
* **`len()`:** This function returns the number of items in the dictionary.
* **`clear()`:** This function removes all elements from the dictionary.
* **`get()`:** This function is used to get the value associated with the key.
* **`pop()`:** This function removes the entry associated with the specified key and returns the corresponding value.
* **`popitem()`:** This function removes an arbitrary item (key-value) from the dictionary and returns it.
* **`keys()`:** This function returns all keys associated with the dictionary.
* **`values()`:** This function returns all values associated with the dictionary.
* **`items()`:** This function returns a list of tuples representing key-value pairs.
* **`copy()`:** This function creates a copy of the dictionary.
* **`setdefault()`:** This function is used to get the value associated with the key. If the key is not available, then the specified key-value will be added as new item to the dictionary.
* **`update()`:** This function updates the dictionary with the items present in the dictionary `x`.

````python
d = {100: "durga", 200: "ravi", 300: "shiva"}
print(len(d))  # 3

d.clear()
print(len(d))  # 0

print(d.get(100))  # durga
print(d.get(400))  # None
print(d.pop(100))  # durga
print(d)  # {200: 'ravi', 300: 'shiva'}

print(d.popitem())  # (300, 'shiva')

print(d)  # {200: 'ravi'}
print(d.keys())  # dict_keys([200])
print(d.values())  # dict_values(['ravi'])
print(d.items())  # [(200, 'ravi')]

d1 = d.copy()
print(d1)  # {200: 'ravi'}
print(d.setdefault(400, "pavan"))  # pavan
print(d)  # {200: 'ravi', 400: 'pavan'}
print(d.setdefault(100, "sachin"))  # durga

print(d)  # {200: 'ravi', 400: 'pavan', 100: 'durga'}
d.update({100: "sunny"})
print(d)  # {100: 'sunny', 200: 'ravi', 400: 'pavan'}
````
